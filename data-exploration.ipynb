```python
def 
find_ip_addr_in_time_range(startdatetime=datetime.datetime(2023,7,25), 
enddatetime=datetime.datetime(2023,7,26)):
    with 
geoip2.database.Reader('/home/daniel/ont/kov09ayc01xbkdkv/GeoLite2-City.mmdb') 
as reader:
        for addr in 
list(client["grouper_ping"]["mux_scan_segment_summary"].find({'logged': 
{"$gte":startdatetime, "$lt":enddatetime}})):
            try:
                response = reader.city(addr["remote_addr"])
                print(response.location.longitude)
                print(response.location.latitude)
                print(response.country.name)
                print(response.country.iso_code)
            except Exception as AddressNotFoundError:
                continue
                
find_ip_addr_in_time_range(datetime.datetime(2023,7,15), 
datetime.datetime(2023,7,16))
```
